```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 5e-3, 5)
```

```python
# Casos de Entrada
vi_caso1 = 10 * np.sin(1000 * t)
vi_caso2 = 1 * np.sin(1000 * t)
```

```python
# Casos de Saída
vo_caso1 = np.tanh(vi_caso1)
vo_caso2 = np.tanh(vi_caso2)
```

```python
# Criando figura com 4 plots
fig, axs = plt.subplots(2, 2, figsize=(15, 12))

#CASO 1: Entrada
axs[0, 0].plot(t, vi_caso1)
axs[0, 0].set_title("Entrada - Caso 1")
axs[0, 0].set_xlabel("t (s)")
axs[0, 0].set_ylabel("Amplitude")
axs[0, 0].grid(True)

#CASO 1: Saída
axs[0, 1].plot(t, vo_caso1)
axs[0, 1].set_title("Saída - Caso 1")
axs[0, 1].set_xlabel("t (s)")
axs[0, 1].set_ylabel("Amplitude")
axs[0, 1].grid(True)

#CASO 2: Entrada
axs[1, 0].plot(t, vi_caso2)
axs[1, 0].set_title("Entrada - Caso 2")
axs[1, 0].set_xlabel("t (s)")
axs[1, 0].set_ylabel("Amplitude")
axs[1, 0].grid(True)

#CASO 2: Saída
axs[1, 1].plot(t, vo_caso2)
axs[1, 1].set_title("Saída - Caso 2")
axs[1, 1].set_xlabel("t (s)")
axs[1, 1].set_ylabel("Amplitude")
axs[1, 1].grid(True)
```
