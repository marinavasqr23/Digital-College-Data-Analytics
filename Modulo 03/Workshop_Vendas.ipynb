```python
import pandas as pd
import glob
```

```python
df = pd.read_excel('/content/vendas_01.xlsx')
```

```python
df.info()
```

```python
arquivos_venda = glob.glob('vendas_*.xlsx')
```

```python
arquivos_venda
```

```python
df_vendas = pd.concat([pd.read_excel(arq) for arq in arquivos_venda],ignore_index=True)
```

```python
df_vendas.head(10)
```

```python
df_vendas[["Empresa", "Filial"]] = df_vendas['Loja'].str.split("-", expand = True)
```

```python
df_vendas
```

```python
df_vendas.drop(columns=['Loja'], inplace=True)
```

```python
df_vendas.head()
```

```python
df_vendas.info()
```

```python
df_vendas['Data'] = pd.to_datetime(df_vendas['Data'])
```

```python
df_vendas.info()
```

```python
df_vendas['Ano'] = df_vendas['Data'].dt.year
```

```python
df_vendas['Mês'] = df_vendas['Data'].dt.month
```

```python
df_vendas.columns
```

```python
df_vendas
```

```python
df_vendas.columns = df_vendas.columns.str.lower()
```

```python
tabela_vendas_filial = df_vendas.groupby(['empresa', 'filial'])['valor (r$)'].sum().reset_index()

```

```python
tabela_vendas_filial.rename(columns={'valor (r$)': 'total filial'}, inplace=True)
```

```python
tabela_vendas_filial
```

```python
tabela_vendas_empresa_geral = df_vendas.groupby(['empresa'])['valor (r$)'].sum().reset_index()
```

```python
tabela_vendas_empresa_geral.rename(columns={'valor (r$)': 'total geral'}, inplace=True)
```

```python
tabela_vendas_empresa_geral
```

```python
tabela_vendas_filial = tabela_vendas_filial.merge(tabela_vendas_empresa_geral, on = 'empresa')
```

```python
tabela_vendas_filial ['Percentual'] = (
    tabela_vendas_filial['total filial']/tabela_vendas_filial['total geral']
)*100
```
