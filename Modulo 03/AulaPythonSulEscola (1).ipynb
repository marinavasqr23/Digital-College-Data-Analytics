```python
dict_replace_bimestre = {
    1: "B1",
    2 : "B2",
    3 : "B3",
    4 : "B4"
}
```

TROCAS DOS VALORES USANDO **MAP**

```python
df['Bimestre'] = df['Bimestre'].map(dict_replace_bimestre)
```

```python
df_turma_1em = df.loc[df["Turma"] == '1EM']
```

```python
df_turma_1em.head(3)
```

**PIVOTAR**: TROCAR LINHAS POR COLUNAS

```python
df_pivot_turma_1em = df_turma_1em.pivot_table(index="Nome", columns=["Materia", "Bimestre"], values="Nota").reset_index()
```

```python
df_pivot_turma_1em
```

```python
materias = ["Ciências Humanas", "Ciências da Natureza", "Linguagens", "Matemática"]
```

```python
df_pivot_turma_1em.head(3)
```

**CALCULANDO DESEMPENHO**

```python
for mat in materias:
    df_pivot_turma_1em[(mat, 'Desempenho B1')] = df_pivot_turma_1em[(mat, 'B1')]
    df_pivot_turma_1em[(mat, 'Desempenho B2')] = (df_pivot_turma_1em[(mat, 'B1')] + df_pivot_turma_1em[(mat, 'B2')])/2
    df_pivot_turma_1em[(mat, 'Desempenho B3')] = (df_pivot_turma_1em[(mat, 'B1')] + df_pivot_turma_1em[(mat, 'B2')] + df_pivot_turma_1em[(mat, 'B3')])/3
    df_pivot_turma_1em[(mat, 'Desempenho B4')] = (df_pivot_turma_1em[(mat, 'B1')] + df_pivot_turma_1em[(mat, 'B2')] + df_pivot_turma_1em[(mat, 'B3')] + df_pivot_turma_1em[(mat, 'B4')])/4
```

```python
df_pivot_turma_1em = df_pivot_turma_1em.sort_index(axis=1, level=0)
```

```python
df_pivot_turma_1em.head()
```

```python
def classificar_nota(valor):
  if 0 < valor <=3:
    return "Ruim"
  elif 3 < valor <=5:
    return "Baixo"
  elif 5 < valor <= 7:
    return "Médio"
  elif 7 < valor <= 8:
    return "Bom"
  elif 8 < valor <= 9:
    return "Ótimo"
  elif 9 < valor <=10:
    return "Excelente"
```

```python
desempenho_bimestre = ["Desempenho B2","Desempenho B2","Desempenho B3","Desempenho B4"]
```

```python
for mat in materias:
  for dem_bi in desempenho_bimestre:
    desem_nota = (mat,dem_bi)
    if desem_nota in df_pivot_turma_1em.columns:
      df_pivot_turma_1em[mat,(f"Status_{dem_bi}")] = df_pivot_turma_1em[(desem_nota)].apply(classificar_nota)
```
